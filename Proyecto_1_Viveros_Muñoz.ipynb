```python
class Directed_graph:
        def __init__(self, nodos, arcos):
           if type(nodos) == list and type(arcos) == list:
                Bool1 = True
                Bool2 = True
                for i in nodos:
                    if type(i) != int:
                        Bool1 = False
                for a in arcos:
                    if type(a) != [str, str, int]:
                        Bool2 = False
           self.nodos = nodos
           self.arcos = arcos

        def antecedentes(self, nodo):
            if nodo not in self.nodos:
                print("El nodo no existe:", nodo)
                return []
            else:
                Ant = []
                for a in self.arcos:
                    if nodo == a[1]:
                        Ant.append(a[0])
                return Ant

        def sucesores(self, nodo):
            if nodo not in self.nodos:
                print("El nodo no existe:", nodo)
                return []
            else:
                Suc = []
                for a in self.arcos:
                    if nodo == a[0]:
                        Suc.append(a[1])
                return Suc

        def add_vertex(self, nodo):
            self.nodos.append(nodo)

        def add_edge(self, arco):
            if type(arco) == list and len(arco) == 3 and arco[0] in self.nodos and arco[1] in self.nodos:
                self.arcos.append(arco)
            else:
                print("Arco inválido.")

        def alcance(self, nodo):
            if nodo not in self.nodos:
                print("El nodo no existe:", nodo)
            else:
                Vec = {}
                for a in self.arcos:
                    if nodo == a[0]:
                        Vec.update({a[1]: a[2]})
                return Vec

        def nodo_inicial(self):
            for nodo in self.nodos:
                if not self.antecedentes(nodo):
                    return nodo
            return "No hay nodo inicial"

        def nodo_final(self):
            for nodo in self.nodos:
                if not self.sucesores(nodo):
                    return nodo
            return "No hay nodo final"

        def get_peso_arco(self, nodo_origen, nodo_destino):
            for arco in self.arcos:
                if arco[0] == nodo_origen and arco[1] == nodo_destino:
                    return arco[2]
            return float('inf')

        #Genera el camino de longitud mínima
        def distancia(self, nodoI, nodoF):
            if nodoI == "No hay nodo inicial" or nodoF == "No hay nodo final"
                print("No se puede encontrar el camino más corto. Falta nodo inicial o final.")
                return [], float('inf')

             dist, prev = {}, {}
             result = []

            for nodo in self.nodos:
                dist[nodo] = float("inf")
                prev[nodo] = None
            dist[nodoI] = 0

            Q = [nodo for nodo in self.nodos]
            #Distancias a los nodos desde el inicio
            while Q:
                u = min(Q, key=dist.get)
                Q.remove(u)
                result.append(u)

                for vecino in self.alcance(u).keys():
                    if vecino in Q and dist[vecino] > dist[u] + self.alcance(u)[vecino]:
                        dist[vecino] = dist[u] + self.alcance(u)[vecino]
                        prev[vecino] = u

            # Reconstruir el camino más corto
            camino = []
            actual = nodoF
            while actual is not None:
                camino.insert(0, actual)
                actual = prev[actual]

            if camino[0] == nodoI:
                return camino
            else:
                return []


    # Leer el archivo Excel
    import pandas as pd
    try:
        df = pd.read_excel("ejemplo.xlsx")
    except:
        print("No se pudo leer el archivo")

    tiempos = {}
    # Crear nodos
    nodos = df[df.columns[1]].tolist()
    nodosIm = df[df.columns[0]].tolist()
    tiempos = {nodo: df.iloc[indice, 3] for indice, nodo in enumerate(nodos)}
    # Crear arcos
    arcos = []
    arcosIm = []
    for indice, fila in df.iterrows():
        if not pd.isnull(fila[df.columns[2]]):
            for num in str(fila[df.columns[2]]).split(','):
                num = num.strip()  # Eliminar espacios en blanco alrededor del valor
                try:
                    indice_actividad = int(float(num)) - 1  # Convertir el valor a número entero
                    if indice_actividad >= 0 and indice_actividad < len(nodos):
                        nodo = nodos[indice_actividad]
                        duracion = tiempos[nodo]
                        arcos.append((nodo, fila[df.columns[1]], duracion))
                        arcosIm.append((indice_actividad+1, fila[df.columns[0]], duracion))
                    else:
                        print("Índice de actividad inválido:", num)
                except ValueError:
                    print("El valor", num, "no es un número válido.")
    # Crear red
    red = Directed_graph(nodos, arcos)

    # Obtener el camino mínimo del nodo inicial final
    camino  = red.distancia(red.nodo_inicial(), red.nodo_final())

    # Escribir el informe en un archivo de texto
    tiempo = 0
    with open("informe.txt", "w") as file:
        file.write("Actividades críticas:n")
        for actividad in camino:
            file.write(f"- {actividad}n")
            tiempo += tiempos[actividad]
        file.write(\"n")
        file.write("Tiempo mínimo:")
        file.write(str(tiempo))
```
